```python
FILENAME_DATASET = 'Dataset/baltimore_crimes.csv'

df = pd.read_csv(FILENAME_DATASET, sep=',', index_col=False)
print (df.head())
locations = df['Location 1'].values
a = []
b = []
coord = []
```

```python
# locations = pd.Series.tolist(locations)
# locations = gmaps.datasets.load_dataset("taxi_rides")
# np.dtype(locations)
for i in range(len(locations)):
#     print (locations[i])
    locations[i] = locations[i].strip('(')
    locations[i] = locations[i].strip(')')
    locations[i] = locations[i].split(',')
    a.append(float(locations[i][0]))
    b.append(float(locations[i][1]))
    temp = (a[i],b[i])
    coord.append(temp)
```

```python
fig = gmaps.figure()
fig.add_layer(gmaps.heatmap_layer(coord))
fig
```

```python
_ = df['District'].value_counts(ascending=True).plot(kind='barh', title='Crime frequency of Districts')
```

```python
day = []
date = df['CrimeDate'].values
dayArray = ['Monday','Tuesday','Wednsday','Thursday','Friday','Saturday','Sunday']
from dateutil.parser import parse
import datetime as dt
for i in range(len(date)):
    temp = dt.datetime.strptime(date[i], '%m/%d/%Y').weekday()
    day.append(dayArray[temp])
```

```python
df['Day'] = day
_ = df['Day'].value_counts(ascending=True).plot(kind='barh', title='Crime frequency of Districts')
```
